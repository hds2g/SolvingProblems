```python
#!/bin/python3

import math
import os
import random
import re
import sys

memo = [0] * 100 #메모이제이션 공간. 전역 변수이므로 0으로 초기화

def fibonacci_memo(n):
    print("a")
    if n<=1:
        return n
    if memo[n] is not 0: #메모가 있는지 확인(0으로 초기화되었으므로 0이 아니라면 메모가 쓰인 것임)
            return memo[n] #메모 리턴
    
    memo[n]=fibonacci_memo(n-1) + fibonacci_memo(n-2) #작은 문제로 분할

    return memo[n]
  
def fibonacci(n):
    print("b")
    if n<=1:
        return n
    ret=fibonacci(n-1) + fibonacci(n-2)
    return ret
```
